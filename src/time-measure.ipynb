```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
```

```python
# Fetching times list
with open('../times.csv', 'r') as file:
    reader = csv.reader(file)
    times = list(reader)
    for row in reader:
        print(row)
times = np.array(times, dtype = np.int32)
times = times[::,0]
times.sort()
#print(times)

# Setting up x axes
step_size = 10000
max_diamonds = 5e5
x_ax = np.zeros(int(max_diamonds / step_size))
for i in range(0, len(x_ax)):
    x_ax[i] = step_size * i

# Setting up aesthetic characteristics
plt.rcParams['font.size']        = 14
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['figure.figsize']   = (15, 10)
plt.xlabel('number of diamonds')
plt.ylabel('clock ticks')

# Actually plotting
plt.plot(x_ax, times)

plt.savefig("../times.png")
```
